### Normalization:
convert to Lower case 

```python

def text_cleaning(text1):
    words=text1.split()                        # divide into words
    re_punct=re.compile('[%s]'%re.escape(string.punctuation))
    w=[re_punct.sub('',w) for w in words]                      # Removing punctuations
    w=[i.lower() for i in w]                                  # Lowering the words
    return (w)
```

**Summarize**

- Load the raw text

- split into tokens

- convert into lowercase

- Remove punctuations from each token

- Filter out each token which are not alphabet

- Filter out tokens that are stop words

```python
# All at one place

import string
import re
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords

# load the data
# load data
df='C:\\Users\\kurre\\OneDrive\\Desktop\\DATA SCIENCE DOCS\\datasets\\metamorphosis clean.txt'
dff=open(df,'rt')
text=dff.read()

# split into tokens
tokens=word_tokenize(text)

#convert into lower case
tokens=[w.lower() for w in tokens]

# prepare regex for removing the punctuations
re_punct=re.compile('[%s]'%re.escape(string.punctuation))

# remove the punctuation
stripped=[re_punct.sub('',w) for w in tokens]

# will remove all tokens that are not alphabetic
words= [word for word in stripped if word.isalpha()]

# remove the stopwords
stop_words=stopwords.words('english')
words=[w for w in words if not w in stop_words]
words[:10]
```

**Lemmatisation**

- Lemmatisation is similar to stemming , as it produces a normalised version of the input word

- The output is lemma i.e.Proper word

- The input word is lemmataised according to its Part-Of-Speech(POS) tag

```python
from nltk.stem import WordNetLemmatizer
s=WordNetLemmatizer()
print(s.lemmatize('having',pos='v'))
print(s.lemmatize('have',pos='v'))
print(s.lemmatize('had',pos='v'))
print('=======================')
print(s.lemmatize('fishing',pos='v'))
print(s.lemmatize('fish',pos='v'))
print(s.lemmatize('fisher',pos='v'))
print(s.lemmatize('fishes',pos='v'))
print(s.lemmatize('fished',pos='v'))
```

```python
from nltk.stem import PorterStemmer
s=PorterStemmer()
print(s.stem("having"))
print(s.stem("have"))
print(s.stem("had"))
```
